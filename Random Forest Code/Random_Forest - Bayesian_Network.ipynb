###Choice of Classifer
Next, we use tree different classifiers to classify the data and test the cross-validated score. We use (1) a binary classification tree, (2) a random forest, and (3) an extra-random forest. The extra random forest is the random forest where a subset of features is examined at each node of the tree. 

```python
#train and validate on raw data

#ONE TREE
clf = DecisionTreeClassifier(max_depth=None, random_state=0)
scores = cross_val_score(clf, X, y)
print "one tree:", scores.mean()

#RANDOM FOREST
clf = RandomForestClassifier(n_estimators=100, n_jobs=-1, random_state=0)
scores = cross_val_score(clf, X, y)
print "Random Forest:", scores.mean()

#EXTRA RANDOM FOREST
clf = ExtraTreesClassifier(n_estimators=100, n_jobs=-1, random_state=0)
scores = cross_val_score(clf, X, y)
print "Extra Random Forest:", scores.mean()
```

It appears that the Extra Random Forest does best.

```python
#normalize
X_norm = np.nan_to_num(sp.stats.mstats.zscore(X,axis=0))

#EXTRA RANDOM FOREST
clf = ExtraTreesClassifier(n_estimators=100, n_jobs=-1, random_state=0)
scores = cross_val_score(clf, X_norm, y)
print "Extra Random Forest:", scores.mean()
```

Normalization has no effect, so it means the SKlearn module does it for us!

###Vary the Number of Trees
We vary the number of trees in the forest, and measure the cross-validation score for each scenario. We expect the best scores to be around the squareroot of the number of data points $(\sqrt{15120}\approx 122)$, just by rule of thumb.

```python
%%time

#EXTRA RANDOM FOREST
n_trees = np.arange(80,200,20)
scs = 0.0*n_trees
for e,n in enumerate(n_trees):
    clf = ExtraTreesClassifier(n_estimators=n, n_jobs=-1, random_state=0)
    scores = cross_val_score(clf, X, y)
    scs[e] = scores.mean()
    print n
```

```python
plt.plot(n_trees,scs)
plt.xlabel('Number of Trees')
plt.ylabel('Cross-Validated Score')
plt.show()
```

It looks like the performance improves with the number of trees up to a certain point. As the graph shows, 122 trees could be a good rule of thumb for estimating the number of trees that would give the best performance. 

###Vary the Number of Features used in Extra Random Fores
Going forward, we use 122 trees in our forest. We next determine a good choice for the number of features to use when splitting trees. We will vary the number of features from 10% to 100% of the 53 features in the data set. 

```python
n_features = np.linspace(0.1,1,8)
scs = 0.0*n_features
for e,f in enumerate(n_features):
    clf = ExtraTreesClassifier(n_estimators=122, max_features=int(f*len(X[0])),  n_jobs=-1, random_state=0)
    scores = cross_val_score(clf, X, y)
    scs[e] = scores.mean()
    print f
```

```python
plt.plot([int(i) for i in n_features*len(X[0])],scs,'.')
plt.xlabel('Number of Features')
plt.ylabel('Cross-Validated Score')
plt.show()
```

It looks like 11, which is close to the square root of 52 works best. We will use 0.22*52 = 11 features going forward, and 122 trees.

To get unbiased random forest classifer scores for each data point, we split the data into thirds, and classify each third with the random forest generated from the other two thirds. 

```python
mask = np.bitwise_or(np.bitwise_or(np.bitwise_or(y==4, y==2), y==5), y==6)
Subset_X = X[mask,:]
Subset_y = y[mask]
Subset_v = np.zeros((len(Subset_y),len(np.unique(Subset_y))))

#shuffle mask
s_mask = np.arange(0,len(Subset_y))
np.random.shuffle(s_mask)
step = len(s_mask)/3

#FIRST THIRD
#get indices we are working with
test_set_mask = s_mask[0:step]
train_set_mask = s_mask[step:]
clf = ExtraTreesClassifier(n_estimators=122, max_features=11) 
clf.fit(Subset_X[train_set_mask],Subset_y[train_set_mask])
Subset_v[test_set_mask,:] = clf.predict_proba(Subset_X[test_set_mask])
print ((Subset_y[test_set_mask] == clf.predict(Subset_X[test_set_mask])).astype(int)-1).sum()/float(step)

#SECOND THIRD
#get indices we are working with
test_set_mask = s_mask[step:2*step]
train_set_mask = np.concatenate((s_mask[:step],s_mask[2*step:]))
clf = ExtraTreesClassifier(n_estimators=122, max_features=11) 
clf.fit(Subset_X[train_set_mask],Subset_y[train_set_mask])
Subset_v[test_set_mask,:] = clf.predict_proba(Subset_X[test_set_mask])
print ((Subset_y[test_set_mask] == clf.predict(Subset_X[test_set_mask])).astype(int)-1).sum()/float(step)

#THIRD THIRD
#get indices we are working with
test_set_mask = s_mask[2*step:]
train_set_mask = s_mask[:2*step]
clf = ExtraTreesClassifier(n_estimators=122, max_features=11) 
clf.fit(Subset_X[train_set_mask],Subset_y[train_set_mask])
Subset_v[test_set_mask,:] = clf.predict_proba(Subset_X[test_set_mask])
print ((Subset_y[test_set_mask] == clf.predict(Subset_X[test_set_mask])).astype(int)-1).sum()/float(step)

#discretize
#Subset_v = np.round_(Subset_v,decimals=1)
```

Now we learn a Bayesian network:

```python
#discretize data
Subset_int = np.round_(Subset_v*4.,decimals=0)

#prep data
bn_data = []
for row in Subset_int:
    temp = {}
    for e,j in enumerate(row):
        temp[e] = j
    bn_data.append(temp)
```

```python
import json

from libpgm.nodedata import NodeData
from libpgm.graphskeleton import GraphSkeleton
from libpgm.lgbayesiannetwork import LGBayesianNetwork
from libpgm.discretebayesiannetwork import DiscreteBayesianNetwork
from libpgm.pgmlearner import PGMLearner
from libpgm.tablecpdfactorization import TableCPDFactorization

# instantiate my learner 
learner = PGMLearner()

# learn bayesian network
result = learner.discrete_estimatebn(bn_data)

# load factorization
result.toporder()
fn = TableCPDFactorization(result)
```

After the network is trained, we want to calculate the joint probability for the points that we used to train the network:

```python
%%time
#calculate joint probability
#Subset_v = np.round(Subset_v,decimals=1)
joint = np.zeros(len(Subset_int))
for e,datapoint in enumerate(Subset_int):
    # learn bayesian network
    fn_temp = copy.deepcopy(fn)
    joint[e] = fn_temp.specificquery({0:[datapoint[0]],1:[datapoint[1]],2:[datapoint[2]],3:[datapoint[3]]},{})
```
