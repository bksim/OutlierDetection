```python
"""
AC 299r
Brandon Sim
"""

import math
import numpy as np
import itertools

# LOCI implementation

""" we can make it more efficient by making a class called a LOCI 
and adding points to it and updating relevant values and pairwise distances 
as new things are added
"""

# gets the r-neighbors of x (all elements e in points such that distance(x, e) <= r
# always contains itself
def neighbors(p, r, points):
    return [a for a in points if distance(p, a) <= r]

# generic distance method
# a, b are n-tuples
# returns the 2-norm
def distance(a, b):
    return math.sqrt(sum((x-y)**2 for x, y in itertools.izip(a, b)))

# max over pi, pj in P of d(pi, pj)
# that is, the greatest distance between any two points in the dataset
def point_set_radius(points):
    r = 0
    for pi in points:
        for pj in points:
            d = distance(pi, pj)
            if d > r:
                r = d
    return r

# finds the m-th nearest neighbor of p in points
def nearest_neighbor(p, m, points):
    """ todo but doesn't seem necessary """
    return

# number of r-neighbors of p
def n(p, r, points):
    return len(neighbors(p, r, points))

# returns nhat and sigma_nhat
# nhat: average of n(p, alpha*r) over the set of r-neighbors of p
# sigma_nhat: standard deviation of n(p, alpha*r) over the set of r-neighbors of p
def nhat(p, r, alpha, points):
    info = {}
    temp = [n(x, alpha*r, points) for x in neighbors(p, r, points)]
    info['nhat'] = np.mean(temp)
    info['sigma_nhat'] = np.std(temp)
    return info

# multi-granularity deviation factor
def MDEF(p, r, alpha, points):
    result = {}
    info = nhat(p, r, alpha, points)
    result['MDEF'] = (1-n(p, alpha*r, points) / info['nhat'])
    result['sigma_MDEF'] = info['sigma_nhat'] / info['nhat']
    return result

# flags outliers according to LOCI algorithm
def LOCI_outliers(points, r, alpha, k_sigma):
    outliers = []
    for p in points:
        mdef = MDEF(p, r, alpha, points)
        if mdef['MDEF'] > k_sigma * mdef['sigma_MDEF']:
            outliers.append(p)
    return outliers
```

```python
""" Let's test our LOCI implementation. Here I download a time series of the SPY, 
an ETF which replicates the S&P 500 Index, from 1993 to present. Then, I will look for
abnormalities in terms of return and volume (ie days in which the return and volume were outliers)
using the LOCI algorithm."""

# get data from quandl
import Quandl
token = "hv4dZPXn5fyuM9CgktzW"
spy = Quandl.get("YAHOO/INDEX_SPY", trim_start="2014-01-01", trim_end="2014-09-29", authtoken=token)
jpy = Quandl.get("YAHOO/INDEX_N225", trim_start="2014-01-01", trim_end="2014-09-29", authtoken=token)
```

```python
# process data from quandl
%matplotlib inline
import matplotlib.pyplot as plt

# dataset: price returns and volumes

#plt.plot(spy['Close'])
#plt.xlabel('Daily close of SPY')
#plt.show()

#plt.plot(spy['Volume'])
#plt.xlabel('Daily volume of SPY')
#plt.show()

daily_returns_original = ((spy['Close'] - spy['Close'].shift())/spy['Close'].shift())[1:]
volume_original = spy['Volume'][0:-1]

jpy_daily_returns_original = ((jpy['Close'] - jpy['Close'].shift())/jpy['Close'].shift())[1:]
jpy_volume_original = jpy['Volume'][0:-1]

# scale values
daily_returns = (daily_returns_original - np.mean(daily_returns_original))/np.std(daily_returns_original)
volume = (volume_original - np.mean(volume_original))/np.std(volume_original)

jpy_daily_returns = (jpy_daily_returns_original - np.mean(jpy_daily_returns_original)+0.01)/np.std(jpy_daily_returns_original)
jpy_volume = (jpy_volume_original - np.mean(jpy_volume_original))/np.std(jpy_volume_original)


daily_returns = list(daily_returns) + list(jpy_daily_returns)
volume = list(volume) + list(jpy_volume)

# only do the last 252 days (approx one year) for speed
#daily_returns = daily_returns[-252:]
#volume = volume[-252:]

# begin LOCI algorithm, may take around a minute or two depending on machine
points = zip(daily_returns, volume)
alpha = 0.5 # default, see paper
k_sigma = 3 # default, see paper

rmax = point_set_radius(itertools.izip(daily_returns, volume)) / alpha
outliers = LOCI_outliers(points, 0.5*rmax, alpha, k_sigma)
```

```python
# plot results
plt.scatter(daily_returns, volume)
plt.xlabel('Daily Return')
plt.ylabel('Volume')
plt.scatter(*zip(*outliers), color='red')
plt.legend(['(Return, Volume) points', 'Outliers'], loc = 'center left', bbox_to_anchor = (0.2, 1.12))
plt.show()
```
