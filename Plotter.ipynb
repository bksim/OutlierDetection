```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.graphics.tsaplots as tsaplots
from  statsmodels.tsa import stattools
import pandas as pd
import FATS as F
import os
# plot some light curves
plt.rcParams['savefig.dpi'] = 300 
```

```python
data_dir = 'Cluster/'
class_name = ['rrlyr', 'rcb']
region = 'V'
lc_dict = {}

plt.figure()

for c in class_name:
    lc_dict[c] = []
    lc_dir = data_dir + c + '/phot/' + region + '/'
    for fn in os.listdir(lc_dir):
        lc_fn = lc_dir + fn
        lc = F.ReadLC_MACHO(lc_fn)
        [data, mjd, error] = lc.ReadLC()
        preprocessed_data = F.Preprocess_LC(data, mjd, error)
        [data, mjd, error] = preprocessed_data.Preprocess()
        if data[0] < 17 and c == 'rrlyr':
            continue
        lc_dict[c] = lc_dict[c] + [(mjd, data, error)]
        plt.plot(mjd, data)
plt.show()
```

```python
plt.figure()
for i in range(len(lc_dict['rrlyr'])):
    plt.plot(lc_dict['rrlyr'][i][0], lc_dict['rrlyr'][i][1], alpha=0.2, color='gray')
    
for j in [0,2,3,4]:
    plt.plot(lc_dict['rcb'][j][0][:-4], lc_dict['rcb'][j][1][:-4], color='red',alpha=0.5)
plt.plot([], label='Typical time series', color='gray', alpha=0.2)
plt.plot([], label='Outlier time series (class rcb)', color='red', alpha=0.5)
plt.legend(bbox_to_anchor=(1.6, 1.05))
plt.xlabel('mjd')
plt.ylabel('Data')
plt.show()
```
