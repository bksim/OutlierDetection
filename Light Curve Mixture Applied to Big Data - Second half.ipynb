```python
def get_data_from_csv(fileregex):
    lc = []
    y = []
    y_id = []
    for filename in glob.glob(fileregex):
        with open(filename, 'r') as f:
            #print filename
            reader = csv.reader(f)
            #skip first row
            reader.next()
            for row in reader:
                lc_row = []
                y_id.append(row[0])
                for r in row[1:]:
                    lc_row.append(r.replace(',','').replace('None','0'))
                lc.append(np.array(lc_row).astype('float'))
                y.append(filename.split('_')[0].split('/')[1])
    return np.nan_to_num(np.array(lc)).astype('float'), np.array(y), np.array(y_id)
```

#Expert 0: KNN

```python
def knn_1(data, r):

    n = len(data)
    k = int(r*n)
    
    data = np.nan_to_num(sp.stats.mstats.zscore(data,axis=0))

    nbrs = NearestNeighbors(n_neighbors=k, algorithm='ball_tree').fit(data)
    distances, indices = nbrs.kneighbors(data)
    
    sum_distances = []
    for d in distances:
        d[d==np.inf] = np.nanmax(d[d!=np.inf])
        sum_distances.append(np.nan_to_num(np.sum(d)))
    
    mean_distance = np.mean(sum_distances)
    sd_distance = np.std(sum_distances)
    
    scores = []
    for i in range(len(sum_distances)):
        scores.append(abs(sum_distances[i] - mean_distance)/sd_distance)
    scores = np.array(scores)
    
    return scores

def knn_2(data, k, d):
    n = len(data)
    data = np.nan_to_num(sp.stats.mstats.zscore(data,axis=0))
    
    nbrs = NearestNeighbors(n_neighbors=k, algorithm='ball_tree').fit(data)
    distances, indices = nbrs.kneighbors(data)
    
    scores = []
    for i in range(len(distances)):
        count = 0
        for a in distances[i]:
            if a < d:
                count += 1
        scores.append(count)
        
    scores = np.array(scores).astype('float')
    
    return sp.stats.norm.pdf( (scores-scores.mean())/scores.std() )
```

```python
def KNN1_wrapper(X_in, X_out, in_y=0, out_y=0, r=0.01):

    #concat points
    points = X_in #np.concatenate((X_in, X_out))
    outlier_mask = np.array([False]*len(X_in) + [True]*len(X_out))
    
    #get scores
    scores = np.nan_to_num(np.array(knn_1(points, r)))

    #HIGHER IS OUTLIER
    return [scores[np.invert(outlier_mask)], scores[outlier_mask]]
```

```python
def KNN2_wrapper(X_in, X_out, in_y=0, out_y=0, k=10, d=25):
    
    
    #concat points
    points = X_in #np.concatenate((X_in, X_out))
    outlier_mask = np.array([False]*len(X_in) + [True]*len(X_out))
    
    #get scores
    scores = np.nan_to_num(np.array(knn_2(points, k, d)))

    #HIGHER IS OUTLIER
    return [scores[np.invert(outlier_mask)], scores[outlier_mask]]
```

#Expert 1 (SVM+JP)

```python
def svm_jp(in_X_data, out_X_data, in_y, out_y, disc_bins=9):
    
    #normalize
    points = in_X_data #np.concatenate((X_in, X_out))
    points = np.nan_to_num(sp.stats.mstats.zscore(points,axis=0))
    in_X_data = points[:len(in_X_data)]
    out_X_data = points[len(in_X_data):]
    
    #create uniform sample
    n_unif = len(in_X_data)
    #create storage for uniform sample space
    X_unif = np.zeros((n_unif, len(in_X_data[0])))
    #choose continuous features
    for cf in range(0,len(in_X_data[0])): #iterate over continuous features, uniformly picking
        X_unif[:,cf] = np.random.uniform(low=in_X_data[:,cf].min()*1.1, high=in_X_data[:,cf].max()*1.1, size=n_unif)
    
    #train SVC
    n_unif_train = len(X_unif)
    X_aug = np.nan_to_num(  np.vstack((in_X_data,X_unif[:n_unif_train]))  )
    y_aug = np.concatenate((in_y,np.array(['Unif']*n_unif_train)))
    rbf_svc = svm.SVC(kernel='rbf', gamma=0.00005, probability=True).fit(X_aug, y_aug)

    #get votes for non-outliers, take only first three probabilites, renormalize
    Z = rbf_svc.predict_proba(in_X_data)
    V = Z[:,:3]/np.repeat([Z[:,:3].sum(axis=1)],3,axis=0).T
    
    #[1] create joint probability table
    #discretize settings
    n_labels = len(V[0])
    #discretize data
    V_disc = np.round_(V*(disc_bins-1),decimals=0)
    #gives unique id to each combination of discretized voting vectors
    flatten_coords = lambda each: int(np.dot(each,np.power(disc_bins, np.arange(1,n_labels +1))))
    #init probability table
    grid = {}
    #for each voting vector, store its freqeuncy
    n_samples = len(V_disc)
    for each in V_disc:
        if flatten_coords(each) not in grid.keys():
            grid[flatten_coords(each)] = 1./n_samples
        else:
            grid[flatten_coords(each)] += 1./n_samples
    
    #[2] apply clf on the outlier data only, discretize, combine with V
    Z_out = rbf_svc.predict_proba(out_X_data)
    out_V_data = Z_out[:,:3]/np.repeat([Z_out[:,:3].sum(axis=1)],3,axis=0).T
    out_V_data_disc = np.round_(out_V_data*(disc_bins-1),decimals=0)

    #[4] lookup joint probability from table
    #for non-outliers
    JP = np.zeros(len(V_disc))
    for en,each in enumerate(V_disc):
        if flatten_coords(each) in grid.keys():
            JP[en] = grid[flatten_coords(each)]
    #for outliers
    out_JP_data = np.zeros(len(out_V_data_disc))
    for en,each in enumerate(out_V_data_disc):
        if flatten_coords(each) in grid.keys():
            out_JP_data[en] = grid[flatten_coords(each)]
    
    #correct signs HIGHER IS OUTLIERS
    return  [1.-np.nan_to_num(JP), 1.-np.nan_to_num(out_JP_data)]
```

#Expert 2 (LOCI)

```python
#LOCI implementation Adpated from Brandom Sim

# returns nhat and sigma_nhat
# nhat: average of n(p, alpha*r) over the set of r-neighbors of p
# sigma_nhat: standard deviation of n(p, alpha*r) over the set of r-neighbors of p
def nhat(p, r, alpha,points, pre_calc_n, KDTree):
    info = {}
    #get number of neighbors of each of the neighbors
    temp = pre_calc_n[KDTree.query_ball_point(p, r)]
    info['nhat'] = temp.mean()
    info['sigma_nhat'] = temp.std()
    return info

# multi-granularity deviation factor
def MDEF(e, p, r, alpha,points, pre_calc_n, KDTree):
    result = {}
    info = nhat(p, r, alpha,points, pre_calc_n, KDTree)
    result['MDEF'] = (1. - pre_calc_n[e] / info['nhat'])
    result['sigma_MDEF'] = info['sigma_nhat']/ info['nhat'] #<<--not sure why this is here
    return result

# calculates outlierliness probability
def LOCI_outliers(points, r, alpha, pre_calc_n, KDTree):
    outliers = np.zeros(len(points))
    for e,p in enumerate(points):
        mdef = MDEF(e,p, r, alpha,points, pre_calc_n, KDTree)
        outliers[e] = sp.stats.norm.pdf((mdef['MDEF'])/mdef['sigma_MDEF'])
    return outliers

def LOCI_wrapper(X_in, X_out,y_in=0, y_out=0, alpha=0.5):
    
    outlier_scores = []
    
    #concat points
    points = X_in #np.concatenate((X_in, X_out))
    scaled_points = np.nan_to_num(sp.stats.mstats.zscore(points, axis=0))
    outlier_mask = np.array([False]*len(X_in) + [True]*len(X_out))
    
    #optimize distance calculations
    KDTree = sp.spatial.cKDTree(scaled_points)
    rmax = sp.spatial.distance.pdist(scaled_points).max() / abs(alpha)
    pre_calc_n = np.ones(len(scaled_points))
    for en,p in enumerate(scaled_points):
        pre_calc_n[en] = len(KDTree.query_ball_point(p, 0.5*abs(alpha)*rmax))

    #get scores
    scores = LOCI_outliers(scaled_points, 0.5*rmax, abs(alpha), pre_calc_n, KDTree)

    #HIGHER IS OUTLIER
    return [scores[np.invert(outlier_mask)], scores[outlier_mask]]
```

#Expert 3 (Eskin)

```python
#modified from WESLEY CHEN
   
def Eskin_outlier_score(lam, points, log_unif_p):
    
    #calculate posterior if all came from normal
    z_norm = sp.stats.mstats.zscore(points, axis=0)
    log_posterior_norm = np.log(1.-lam) + np.sum(np.nan_to_num(sp.stats.norm.logpdf(z_norm)),axis=1)
    
    #calculate posterior if all came from uniform
    log_post_outlier = np.log(lam) - log_unif_p

    return log_post_outlier - log_posterior_norm

def Eskin_wrapper(X_in, X_out, y_in=0, y_out=0, lam=0.5):
    
    #concat points
    points = X_in #np.concatenate((X_in, X_out))
    outlier_mask = np.array([False]*len(X_in) + [True]*len(X_out))
    
    #get log_unif_p
    ranges = (np.max(X_in,axis=0) - np.min(X_in,axis=0))
    log_unif_p = np.log(ranges[ranges!=0.]).sum()
    
    #get scores
    scores = Eskin_outlier_score(lam, points, log_unif_p)

    #HIGHER IS OUTLIER
    return [scores[np.invert(outlier_mask)], scores[outlier_mask]]
```

#Expert 4 (Hyperplane Regression)

```python
# from RegressiongMethod/Hyperplane Regression.ipynb second to last cell

def hyperreg(data, r):
    # cast as np array for indexing
    data = np.nan_to_num(sp.stats.mstats.zscore(data,axis=0)) 

    X = data[:,1:]
    y = data[:,0]
    # add constants to get a constant weight term
    X = np.hstack((X,np.ones((len(X),1))))

    coeffs, _, _, _ = np.linalg.lstsq(X,y)
    fullNormCoeff = np.linalg.norm(coeffs)
    
    # repeat with removing each row
    diffScore = []
    
    for i in xrange(len(data)):
        
        mask = np.ones(len(data),dtype=bool)
        
        # remove one point at a time
        mask[i] = 0
        myData = data[mask]
        
        X = myData[:,1:]
        y = myData[:,0]

        # add constants to get a constant weight term
        X = np.hstack((X,np.ones((len(X),1))))
    
        coeffs, _, _, _ = np.linalg.lstsq(X,y)
        myNormCoeff = np.linalg.norm(coeffs)
        
        # add to diffscore
        diffScore.append(np.abs(myNormCoeff-fullNormCoeff))
        
    # convert to z-score to report like earlier format
    return np.nan_to_num(np.abs(sp.stats.mstats.zscore(diffScore)))
```

```python
def hyperreg_wrapper(X_in, X_out, in_y=0, out_y=0, r=0.1):
    
    #concat points
    points = X_in #np.concatenate((X_in, X_out))
    points = np.nan_to_num(sp.stats.mstats.zscore(points, axis=0))
    outlier_mask = np.array([False]*len(X_in) + [True]*len(X_out))
    
    #get scores
    scores = np.nan_to_num(np.array(hyperreg(points, r)))

    #HIGHER IS OUTLIER
    # I think we don't need the sign correction here right?
    return [scores[np.invert(outlier_mask)], scores[outlier_mask]]
```

```python
def ROC_curve(eta_number, legend_mod=""):
    #set eta with optimized eta from other ipython notebook
    moo = pickle.load(open('moo_r'+str(eta_number)+'.pk', 'rb'))
    
    #calculate g for each point
    eta_I = moo[:,:n_exp/2]/np.repeat([np.sum(np.abs(moo[:,:n_exp/2]),axis=0)],len(moo[:,:n_exp/2]), axis=0)
    ndotx = np.exp(np.dot(X_comb_I,eta_I))
    ndotx[ndotx == np.inf] = 1.0e10
    ndotx[ndotx == -np.inf] = -1.0e10
    g_I = ndotx/np.repeat([np.sum(ndotx,axis=1)],n_exp/2, axis=0).T

    eta_V = moo[:,n_exp/2:]/np.repeat([np.sum(np.abs(moo[:,n_exp/2:]),axis=0)],len(moo[:,n_exp/2:]), axis=0)
    ndotx = np.exp(np.dot(X_comb_V,eta_V))
    ndotx[ndotx == np.inf] = 1.0
    ndotx[ndotx == -np.inf] = -1.0
    g_V = ndotx/np.repeat([np.sum(ndotx,axis=1)],n_exp/2, axis=0).T

    g = np.hstack((g_I,g_V))
    
    normalized = (g*np.nan_to_num(sp.stats.mstats.zscore(Scores, axis=0))).sum(axis=1)

    X_ingroup = normalized[y_I!='rcb']
    X_outliers = normalized[y_I=='rcb']
    skip = 70

    #define truth and weights
    y_true = np.array([0]*len(X_ingroup)+ [1]*len(X_outliers))

    #get scores
    y_scores = np.concatenate((X_ingroup,X_outliers))

    #plot ROC
    fpr, tpr, thresh = roc_curve(y_true, y_scores)
    AUC = round(np.sum(  np.diff(fpr)*0.5*(tpr[1:]+tpr[:-1])  ),2)
    plt.plot(fpr, tpr, '.-', label="Detection of RCB class as outliers, eta "+str(legend_mod)+", AUC="+str(AUC))

    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.legend(loc=4)
    plt.title('Mixture of Experts Outlier Detection on Large Dataset')
```
