```python
"""
Eskin 2000 Paper Functions

"""

# can generalize with custom probability density functions?

# for now assume non-outliers follow ~Norm
def Pnormal(x, y, XdimStats, YdimStats):
    xprob = (abs(x-XdimStats[0])/XdimStats[1] )/XdimStats[2]
    #xprob = scistat.norm.cdf((abs(x-XdimStats[0])/XdimStats[1] ))
    yprob = (abs(y-YdimStats[0])/YdimStats[1] )/YdimStats[2]
    #yprob = scistat.norm.cdf((abs(y-YdimStats[0])/YdimStats[1] ))
    #print "xnorm"
    #print xprob    
    #print "ynorm"
    #print yprob
    return xprob * yprob

# for now assume outiers are ~Unif
def Poutlier(x, y, XdimStats, YdimStats):
    xprob = 1 / (XdimStats[3] - XdimStats[2])
    yprob = 1 / (YdimStats[3] - YdimStats[2])
    #print "xprobunif"
    #print xprob
    #print "yprobunif"
    #print yprob
    return xprob * yprob
    
def computeLogLikelihood(normpoints, outliers, XdimStats, YdimStats):
    # lambda for probability of being an outlier at random (how many there are)
    myLambda = 0.1;
    
    #print "normloop"
    LLnorm = np.log((1-myLambda))*len(normpoints)
    for normp in normpoints:
        #print "next point"
        #print normp
        LLnorm = LLnorm + np.log(Pnormal(normp[0], normp[1], XdimStats, YdimStats))
    
    #print "outlierloop"
    LLoutlier = np.log(myLambda)*len(outliers)
    for outlier in outliers:
        #print outlier
        LLoutlier = LLoutlier + np.log(Poutlier(outlier[0], outlier[1], XdimStats, YdimStats))
        
    #print "joint prob"
    #print LLnorm
    #print LLoutlier
        
    return LLnorm + LLoutlier
    

def detect2DOutliers(allpoints, XdimStats, YdimStats, thresh):
    
    # start with no outliers and all normal points
    
    # do we need deepcopy? does this optimize to copy if we need it
    normpoints = copy.deepcopy(allpoints)
    outliers = []
    
    # iterate through each point
    for p in allpoints:
        
        # if we add the current point to our outlier subset and remove from the normal set
        normWithoutP = copy.deepcopy(normpoints)
        normWithoutP.remove(p)
        outliersWithP = copy.deepcopy(outliers)
        outliersWithP.append(p)
        
        # compute scores for the two cases
        asNormScore = computeLogLikelihood(normpoints, outliers, XdimStats, YdimStats)
        asOutScore = computeLogLikelihood(normWithoutP, outliersWithP, XdimStats, YdimStats)
        
        #print "normScore"
        #print asNormScore
        #print "outScore"
        #print asOutScore
        #print "diff"
        #print asNormScore - asOutScore
        
        # if the difference is over a threshold, add to outliers
        if (asNormScore - asOutScore) > thresh:
            outliers.append(p);
            normpoints.remove(p)
            
    return outliers
```

```python
# dataset: price returns and volumes

daily_returns_original = ((spy['Close'] - spy['Close'].shift())/spy['Close'].shift())[1:]
volume_original = spy['Volume'][0:-1]

#jpy_daily_returns_original = ((jpy['Close'] - jpy['Close'].shift())/jpy['Close'].shift())[1:]
#jpy_volume_original = jpy['Volume'][0:-1]

#daily_returns = list(daily_returns) + list(jpy_daily_returns)
#volume = list(volume) + list(jpy_volume)

nDaysBack = 250
# only do the last 265 days (YTD) or speed
#daily_returns = daily_returns_original[-1*nDaysBack:]
#volume = volume_original[-1*nDaysBack:]

daily_returns = daily_returns_original[0:-1]
volume = volume_original[0:-1]


# normalize to subset
daily_returns = (daily_returns - np.mean(daily_returns))/np.std(daily_returns)
volume = (volume - np.mean(volume))/np.std(volume)

#jpy_daily_returns = (jpy_daily_returns - np.mean(jpy_daily_returns)+0.01)/np.std(jpy_daily_returns)
#jpy_volume = (jpy_volume - np.mean(jpy_volume))/np.std(jpy_volume)

# zip data 
points = zip(daily_returns, volume)

# should already be mean 0, std 1, because we have pre-normalized
returnStats = [np.mean(daily_returns), np.std(daily_returns), np.min(daily_returns), np.max(daily_returns)]
volStats = [np.mean(volume), np.std(volume), np.min(volume), np.max(volume)]

#print returnStats
#print volStats

# threshold to adjust, higher is less sensitive
thresh = 4.6

# get outliers via Eskin method
outliers = detect2DOutliers(points, returnStats, volStats, thresh)
```

```python
# plot results
plt.scatter(daily_returns, volume)
plt.xlabel('Daily Return')
plt.ylabel('Volume')
plt.scatter(*zip(*outliers), color='red')
plt.legend(['(Return, Volume) points', 'Outliers'], loc = 'center left', bbox_to_anchor = (0.2, 1.12))
plt.show()
```
