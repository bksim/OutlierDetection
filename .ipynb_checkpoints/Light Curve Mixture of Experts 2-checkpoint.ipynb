```python
def get_data_from_csv(fileregex):
    lc = []
    y = []
    y_id = []
    for filename in glob.glob(fileregex):
        with open(filename, 'r') as f:
            reader = csv.reader(f)
            #skip first row
            reader.next()
            for row in reader:
                lc_row = []
                y_id.append(row[0])
                for r in row[1:]:
                    lc_row.append(r.replace(',','').replace('None','0'))
                lc.append(np.array(lc_row).astype('float'))
                y.append(filename.split('_')[0].split('/')[1])
    return np.nan_to_num(np.array(lc)).astype('float'), np.array(y), np.array(y_id)
```

```python
#creates artificail outliers
def artificial_outliers(X_full, out_n=10):

    #bootstrap a sample of out_n outliers
    indices = range(0,len(X_full))
    X_outliers_pre = X_full[np.random.choice(indices,out_n)]

    #randomize continuous features
    for temp in range(0,100): #shuffle 100 times
        for cf in range(0,len(X_full[0])): #iterate over continuous features
            np.random.shuffle(X_outliers_pre[:,cf])

    for cf in range(0,len(X_full[0])): #iterate over continuous features, and add standard deviations away from mean
        X_outliers_pre[:,cf]+(np.random.rand(out_n)+4)*X_full[:,cf].std()
        
    return X_outliers_pre
```

```python
def PCA_plot(X_full, y, X_Outliers):

    #plt.figure(figsize=(15,7))
    out_n = len(X_Outliers)
    
    mypca = PCA()
    #normalize X_full
    X_normed = np.nan_to_num(sp.stats.mstats.zscore(X_full,axis=0))
    mypca.fit(X_normed)
    X_pca_full = mypca.transform(X_normed)
    #take first 2 PCs
    X_pca = X_pca_full[:,:2]
    
    #normalize outliers, plot
    X_normed_X = (X_Outliers - (np.repeat([np.mean(X_full,axis=0)],out_n,axis=0)))/(np.repeat([np.std(X_full,axis=0)],
                                                                                              out_n,axis=0))
    X_pca_X = mypca.transform(np.nan_to_num(X_normed_X))
    #plt.scatter(X_pca_X[:,0] ,X_pca_X[:,1],c = 'w',alpha=1, label="Artificial Outliers", linewidth=2)

    #plot X_full
    y_classes = uniq(y)
    for en, yi in enumerate(y_classes):
        colors='cmykgbrcmykgbr'
        X_current_plot = X_pca[y==yi,:]
        plt.scatter(X_current_plot[:,0],X_current_plot[:,1], c=colors[en],
                    alpha=0.3, linewidth=0, label=yi, s=8)
    
    plt.legend(loc=3)
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    #plt.show()
```

```python
def plot_single_expert(exp_wrapper):

    plt.figure(figsize=(20,5))

    #plot for artificial outliers, I
    X_ingroup = np.nan_to_num(lc_I)
    X_outliers = np.nan_to_num(X_out_I)
    in_y = y_I
    out_y = np.repeat('Art',len(X_out_I))
    
    plt.subplot(1,2,1)
    plt.hist(exp_wrapper(X_ingroup, X_outliers, in_y, out_y), normed = True)
    plt.xlabel('Score (Normal -----> Outliers)')
    plt.ylabel('Frequency')
    plt.legend(['Normal','Outliers'])
    plt.title('I')

    #plot for artificial outliers, V
    X_ingroup = np.nan_to_num(lc_V)
    X_outliers = np.nan_to_num(X_out_V)
    in_y = y_V
    out_y = np.repeat('Art',len(X_out_V))
    
    plt.subplot(1,2,2)
    plt.hist(exp_wrapper(X_ingroup, X_outliers, in_y, out_y), normed = True)
    plt.xlabel('Score (Normal -----> Outliers)')
    plt.ylabel('Frequency')
    plt.legend(['Normal','Outliers'])
    plt.title('V')

    plt.show()
```

```python
def each_ROC_curve(name,exp_wrapper,X_ingroup,X_outliers, in_y, out_y, skip):
    
    #define truth and weights
    y_true = np.array([0]*len(X_ingroup)+ [1]*len(X_outliers))
    y_weight = np.array([1./len(X_ingroup)]*len(X_ingroup)+ [1./len(X_outliers)]*len(X_outliers))
    
    #get scores
    [in_JP,out_JP] = exp_wrapper(X_ingroup, X_outliers, in_y, out_y)
    y_scores = np.concatenate((in_JP,out_JP))
    
    #plot ROC
    fpr, tpr, thresh = roc_curve(y_true, y_scores, sample_weight=y_weight)
    plt.plot(fpr, tpr, '.-', label=name)
    
    #plot thresholds
    for xx,yy,t in zip(fpr[::skip], tpr[::skip],thresh[::skip]):
        plt.text(xx, yy, round(t,5), fontsize=10, rotation =-45)
```

```python
def ROC_curve(exp_wrapper, skip = 1, title=''):
    
    plt.figure(figsize=(20,5))
    
    #plot for artificial outliers, I
    X_ingroup = np.nan_to_num(lc_I)
    X_outliers = np.nan_to_num(X_out_I)
    in_y = y_I
    out_y = np.repeat('Art',len(X_out_I))
    
    plt.subplot(1,2,1)
    each_ROC_curve('Detection of artificial outliers',exp_wrapper,X_ingroup,X_outliers,in_y, out_y,skip)

    #plot for ingroup outliers, I
    y_classes = ['rcb']
    for en, yi in enumerate(y_classes):
    
        X_ingroup = np.nan_to_num(lc_I[y_I!=yi])
        X_outliers = np.nan_to_num(lc_I[y_I==yi])
        in_y = y_I[y_I!=yi]
        out_y = y_I[y_I==yi]
        
        plt.subplot(1,2,1)
        each_ROC_curve('Detection of RCB class as outliers',exp_wrapper,X_ingroup,X_outliers,in_y, out_y,skip)

    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.legend(loc=4)
    plt.title(title+' (I Channel)')
    
    #plot for artificial outliers, V
    X_ingroup = np.nan_to_num(lc_V)
    X_outliers = np.nan_to_num(X_out_V)
    in_y = y_V
    out_y = np.repeat('Art',len(X_out_V))
    
    plt.subplot(1,2,2)
    each_ROC_curve('Detection of artificial outliers',exp_wrapper,X_ingroup,X_outliers,in_y, out_y,skip)

    #plot for ingroup outliers, V
    y_classes = ['rcb']
    for en, yi in enumerate(y_classes):
    
        X_ingroup = np.nan_to_num(lc_V[y_V!=yi])
        X_outliers = np.nan_to_num(lc_V[y_V==yi])
        in_y = y_V[y_V!=yi]
        out_y = y_V[y_V==yi]

        plt.subplot(1,2,2)
        each_ROC_curve('Detection of RCB class as outliers',exp_wrapper,X_ingroup,X_outliers,in_y, out_y,skip)

    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.legend(loc=4)
    plt.title(title+' (V Channel)')

    plt.show()
```

```python
from sklearn.neighbors import NearestNeighbors
def knn_1(data, r):
    n = len(data)
    k = int(r*n)
    
    data = np.nan_to_num(data)

    nbrs = NearestNeighbors(n_neighbors=k, algorithm='ball_tree').fit(data)
    distances, indices = nbrs.kneighbors(data)
    
    sum_distances = []
    for d in distances:
        d[d==np.inf] = np.nanmax(d[d!=np.inf])
        sum_distances.append(np.nan_to_num(np.sum(d)))
    
    mean_distance = np.mean(sum_distances)
    sd_distance = np.std(sum_distances)
    
    scores = []
    for i in range(len(sum_distances)):
        scores.append(abs(sum_distances[i] - mean_distance)/sd_distance)
    scores = np.array(scores)
    
    return sp.stats.norm.pdf( (scores-scores.mean())/scores.std() )

def knn_2(data, k, d):
    n = len(data)
    data = np.nan_to_num(data)
    
    nbrs = NearestNeighbors(n_neighbors=k, algorithm='ball_tree').fit(data)
    distances, indices = nbrs.kneighbors(data)
    
    scores = []
    for i in range(len(distances)):
        count = 0
        for a in distances[i]:
            if a < d:
                count += 1
        scores.append(count)
        
    scores = np.array(scores).astype('float')
    
    return sp.stats.norm.pdf( (scores-scores.mean())/scores.std() )
```

```python
def KNN1_wrapper(X_in, X_out, in_y=0, out_y=0, r=0.1):
    
    #concat points
    points = np.concatenate((X_in, X_out))
    outlier_mask = np.array([False]*len(X_in) + [True]*len(X_out))
    
    #get scores
    scores = np.nan_to_num(np.array(knn_1(points, r)))

    #HIGHER IS OUTLIER
    return [1-scores[np.invert(outlier_mask)], 1-scores[outlier_mask]]
```

```python
def KNN2_wrapper(X_in, X_out, in_y=0, out_y=0, k=10, d=25):
    
    #concat points
    points = np.concatenate((X_in, X_out))
    outlier_mask = np.array([False]*len(X_in) + [True]*len(X_out))
    
    #get scores
    scores = np.nan_to_num(np.array(knn_2(points, k, d)))

    #LOWER IS OUTLIER, so need to correct signs
    return [-scores[np.invert(outlier_mask)], -scores[outlier_mask]]
```

#Expert 1 (SVM+JP)

```python
def svm_jp(in_X_data, out_X_data, in_y, out_y, disc_bins=9):
    
    #create uniform sample
    n_unif = len(in_X_data)
    #create storage for uniform sample space
    X_unif = np.zeros((n_unif, len(in_X_data[0])))
    #choose continuous features
    for cf in range(0,len(in_X_data[0])): #iterate over continuous features, uniformly picking
        X_unif[:,cf] = np.random.uniform(low=in_X_data[:,cf].min()*1.1, high=in_X_data[:,cf].max()*1.1, size=n_unif)
    
    #train SVC
    n_unif_train = len(X_unif)
    X_aug = np.nan_to_num(  np.vstack((in_X_data,X_unif[:n_unif_train]))  )
    y_aug = np.concatenate((in_y,np.array(['Unif']*n_unif_train)))
    rbf_svc = svm.SVC(kernel='rbf', gamma=0.00005, probability=True).fit(X_aug, y_aug)

    #get votes for non-outliers, take only first three probabilites, renormalize
    Z = rbf_svc.predict_proba(in_X_data)
    V = Z[:,:3]/np.repeat([Z[:,:3].sum(axis=1)],3,axis=0).T
    
    #[1] create joint probability table
    #discretize settings
    n_labels = len(V[0])
    #discretize data
    V_disc = np.round_(V*(disc_bins-1),decimals=0)
    #gives unique id to each combination of discretized voting vectors
    flatten_coords = lambda each: int(np.dot(each,np.power(disc_bins, np.arange(1,n_labels +1))))
    #init probability table
    grid = {}
    #for each voting vector, store its freqeuncy
    n_samples = len(V_disc)
    for each in V_disc:
        if flatten_coords(each) not in grid.keys():
            grid[flatten_coords(each)] = 1./n_samples
        else:
            grid[flatten_coords(each)] += 1./n_samples
    
    #[2] apply clf on the outlier data only, discretize, combine with V
    Z_out = rbf_svc.predict_proba(out_X_data)
    out_V_data = Z_out[:,:3]/np.repeat([Z_out[:,:3].sum(axis=1)],3,axis=0).T
    out_V_data_disc = np.round_(out_V_data*(disc_bins-1),decimals=0)

    #[4] lookup joint probability from table
    #for non-outliers
    JP = np.zeros(len(V_disc))
    for en,each in enumerate(V_disc):
        if flatten_coords(each) in grid.keys():
            JP[en] = grid[flatten_coords(each)]
    #for outliers
    out_JP_data = np.zeros(len(out_V_data_disc))
    for en,each in enumerate(out_V_data_disc):
        if flatten_coords(each) in grid.keys():
            out_JP_data[en] = grid[flatten_coords(each)]
    
    #correct signs HIGHER IS OUTLIERS
    return  [1.-np.nan_to_num(JP), 1.-np.nan_to_num(out_JP_data)]
```

#Expert 2 (LOCI)

```python
#LOCI implementation Adpated from Brandom Sim

# returns nhat and sigma_nhat
# nhat: average of n(p, alpha*r) over the set of r-neighbors of p
# sigma_nhat: standard deviation of n(p, alpha*r) over the set of r-neighbors of p
def nhat(p, r, alpha,points, pre_calc_n, KDTree):
    info = {}
    #get number of neighbors of each of the neighbors
    temp = pre_calc_n[KDTree.query_ball_point(p, r)]
    info['nhat'] = temp.mean()
    info['sigma_nhat'] = temp.std()
    return info

# multi-granularity deviation factor
def MDEF(e, p, r, alpha,points, pre_calc_n, KDTree):
    result = {}
    info = nhat(p, r, alpha,points, pre_calc_n, KDTree)
    result['MDEF'] = (1. - pre_calc_n[e] / info['nhat'])
    result['sigma_MDEF'] = info['sigma_nhat']/ info['nhat'] #<<--not sure why this is here
    return result

# calculates outlierliness probability
def LOCI_outliers(points, r, alpha, pre_calc_n, KDTree):
    outliers = np.zeros(len(points))
    for e,p in enumerate(points):
        mdef = MDEF(e,p, r, alpha,points, pre_calc_n, KDTree)
        outliers[e] = sp.stats.norm.pdf((mdef['MDEF'])/mdef['sigma_MDEF'])
    return outliers

def LOCI_wrapper(X_in, X_out,y_in=0, y_out=0, alpha=0.5):
    
    #scale points
    points = np.concatenate((X_in, X_out))
    outlier_mask = np.array([False]*len(X_in) + [True]*len(X_out))
    scaled_points = np.nan_to_num(sp.stats.mstats.zscore(points, axis=0))
    
    #optimize distance calculations
    KDTree = sp.spatial.cKDTree(scaled_points)
    rmax = sp.spatial.distance.pdist(scaled_points).max() / abs(alpha)
    
    #pre-calculate n, number of neighbors
    pre_calc_n = np.ones(len(scaled_points))
    for en,p in enumerate(scaled_points):
        pre_calc_n[en] = len(KDTree.query_ball_point(p, 0.5*abs(alpha)*rmax))
    scores = LOCI_outliers(scaled_points, 0.5*rmax, abs(alpha), pre_calc_n, KDTree)
    
    
    #HIGHER IS OUTLIER
    return [scores[np.invert(outlier_mask)], scores[outlier_mask]] 
```

#Expert 3 (Eskin)

```python
#modified from WESLEY CHEN
   
def Eskin_outlier_score(lam, points, log_unif_p):
    
    #calculate posterior if all came from normal
    z_norm = sp.stats.mstats.zscore(points, axis=0)
    log_posterior_norm = np.log(1.-lam) + np.sum(np.nan_to_num(sp.stats.norm.logpdf(z_norm)),axis=1)
    
    #calculate posterior if all came from uniform
    log_post_outlier = np.log(lam) - log_unif_p

    return log_post_outlier - log_posterior_norm

def Eskin_wrapper(X_in, X_out, y_in=0, y_out=0, lam=0.5):
    
    #concat points
    points = np.concatenate((X_in, X_out))
    outlier_mask = np.array([False]*len(X_in) + [True]*len(X_out))
    
    #get log_unif_p
    ranges = (np.max(X_in,axis=0) - np.min(X_in,axis=0))
    log_unif_p = np.log(ranges[ranges!=0.]).sum()
    
    #get scores
    scores = Eskin_outlier_score(lam, points, log_unif_p)

    #HIGHER IS OUTLIER
    return [scores[np.invert(outlier_mask)], scores[outlier_mask]]
```

```python
# from RegressiongMethod/Hyperplane Regression.ipynb second to last cell

def hyperreg(data, r):
    # cast as np array for indexing
    data = np.array(data)
    n = len(data)
    numpoints = len(data[0])
    data = np.nan_to_num(data)    
    
    X = data[:,1:n]
    y = data[:,0]
    
    # add constants to get a constant weight term
    X = sm.add_constant(X)
    
    # fit a OLS model with the 1:n features as dependent variables
    est = sm.OLS(y, X).fit()
    
    # get norm of coefficients
    fullNormCoeff = np.lingalg.norm(est.params)
    
    # repeat with removing each row
    diffScore = []
    
    for i in xrange(n):
        
        mask = np.ones(numpoints,dtype=bool)
        
        # remove one point at a time
        mask[i] = 0
        myData = data[mask]
        
        X = myData[:,1:n]
        y = myData[:,0]

        # add constants to get a constant weight term
        X = sm.add_constant(X)
    
        # fit a OLS model with the 1:n features as dependent variables
        est = sm.OLS(y, X).fit()
    
        # get norm of coefficients
        myNormCoeff = np.lingalg.norm(est.params)
        
        # add to diffscore
        
        diffscore.append(np.abs(myNormCoeff-fullNormCoeff))
    
    # as per earlier formats like KNN1
    meanDiffScore = np.mean(diffScore)
    sdDiffScore = np.std(diffScore)
    
    scores = []
    for i in range(len(diffScore)):
        scores.append(abs(diffScore[i] - meanDiffScore)/sdDiffScore)
        
    # convert to z-score to report like earlier format
    scores = np.array(scores)
    return sp.stats.norm.pdf( (scores-scores.mean())/scores.std() )
```

```python
n_exp = 10

Scores = np.zeros((len(lc_I)+len(X_out_I), n_exp))

experts = [KNN1_wrapper, KNN2_wrapper, svm_jp, LOCI_wrapper, Eskin_wrapper]

#iterate over I and V

for e,expert in enumerate(experts):
    Scores[:len(lc_I),e], Scores[len(lc_I):,e] = expert(lc_I, X_out_I, y_I, np.repeat('Art',len(X_out_I)))
    
for e,expert in enumerate(experts):
    Scores[:len(lc_I),e+n_exp/2], Scores[len(lc_I):,e+n_exp/2] = expert(lc_V, X_out_V, y_V, np.repeat('Art',len(X_out_V)))
```

```python
#stack X
X_comb_I = np.nan_to_num(sp.stats.mstats.zscore( np.vstack((lc_I,X_out_I)) , axis=0))
X_comb_V = np.nan_to_num(sp.stats.mstats.zscore( np.vstack((lc_V,X_out_V)) , axis=0))

#define function to MINIMIZE
#eta is 3x10 matrix
def objective_function(eta):

    #force the correct shape
    eta = eta.reshape((len(X_comb[0]), n_exp))
    
    #calculate g for each point
    eta_I = eta[:,:n_exp/2]/np.repeat([np.sum(np.abs(eta[:,:n_exp/2]),axis=0)],len(eta[:,:n_exp/2]), axis=0)
    ndotx = np.exp(np.dot(X_comb_I,eta_I))
    ndotx[ndotx == np.inf] = 1.0e10
    ndotx[ndotx == -np.inf] = -1.0e10
    g_I = ndotx/np.repeat([np.sum(ndotx,axis=1)],n_exp/2, axis=0).T
    
    eta_V = eta[:,n_exp/2:]/np.repeat([np.sum(np.abs(eta[:,n_exp/2:]),axis=0)],len(eta[:,n_exp/2:]), axis=0)
    ndotx = np.exp(np.dot(X_comb_V,eta_V))
    ndotx[ndotx == np.inf] = 1.0e10
    ndotx[ndotx == -np.inf] = -1.0e10
    g_V = ndotx/np.repeat([np.sum(ndotx,axis=1)],n_exp/2, axis=0).T
    
    g = np.hstack((g_I,g_V))
    
    #weight scores
    normalized = (g*np.nan_to_num(sp.stats.mstats.zscore(Scores, axis=0))).sum(axis=1)
    Normals = normalized[:len(lc_I)]
    Outgroups = normalized[len(lc_I):]
    
    #optimize both!
    output = np.nan_to_num( np.mean(Normals[y_I!='rcb']) - np.mean(Outgroups) - np.mean(Normals[y_I=='rcb']))
    
    return output
```

```python
#init eta
eta = np.random.normal(size=(len(X_comb[0]), n_exp))

#minimize!
resbrute = sp.optimize.basinhopping(objective_function, eta, niter=5 )
```

```python
#extract out optimized eta
moo = resbrute['x'].reshape((len(X_comb[0]), n_exp))
moo = moo/np.repeat([np.sum(np.abs(moo),axis=0)],len(moo), axis=0)

#calculate g for each point
eta_I = moo[:,:n_exp/2]/np.repeat([np.sum(np.abs(moo[:,:n_exp/2]),axis=0)],len(moo[:,:n_exp/2]), axis=0)
ndotx = np.exp(np.dot(X_comb_I,eta_I))
ndotx[ndotx == np.inf] = 1.0e10
ndotx[ndotx == -np.inf] = -1.0e10
g_I = ndotx/np.repeat([np.sum(ndotx,axis=1)],n_exp/2, axis=0).T

eta_V = moo[:,n_exp/2:]/np.repeat([np.sum(np.abs(moo[:,n_exp/2:]),axis=0)],len(moo[:,n_exp/2:]), axis=0)
ndotx = np.exp(np.dot(X_comb_V,eta_V))
ndotx[ndotx == np.inf] = 1.0
ndotx[ndotx == -np.inf] = -1.0
g_V = ndotx/np.repeat([np.sum(ndotx,axis=1)],n_exp/2, axis=0).T

g = np.hstack((g_I,g_V))

#hitogram of gating
plt.hist(np.nan_to_num(g), bins=10)
plt.legend(['KNN2_I', 'KNN2_I','Pavlos_I','LOCI_I','Eskin_I','KNN2_V', 'KNN2_V','Pavlos_V','LOCI_V','Eskin_V'])
plt.xlabel('gating weights')
plt.ylabel('number of points')
plt.show()
```
