```python
# imports
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# get data from quandl
import Quandl
token = "hv4dZPXn5fyuM9CgktzW"
spy = Quandl.get("YAHOO/INDEX_SPY", trim_start="2014-01-01", trim_end="2014-11-07", authtoken=token)
```

```python
daily_returns_original = ((spy['Close'] - spy['Close'].shift())/spy['Close'].shift())[1:]
volume_original = spy['Volume'][0:-1]
# scale values
daily_returns = (daily_returns_original - np.mean(daily_returns_original))/np.std(daily_returns_original)
volume = (volume_original - np.mean(volume_original))/np.std(volume_original)
```

```python
# plot original dataset
plt.scatter(daily_returns, volume)
plt.xlabel('Daily Return')
plt.ylabel('Volume')
#plt.scatter(*zip(*outliers), color='red')
#plt.legend(['(Return, Volume) points', 'Outliers'], loc = 'center left', bbox_to_anchor = (0.2, 1.12))
plt.grid()
plt.title('Scaled daily volume vs. Scaled daily return for SPY')
plt.show()
```

```python
data = np.array(zip(daily_returns, volume))

# k-nearest neighbor algo 1
from sklearn.neighbors import NearestNeighbors
def knn_1(data, r, t):
    n = len(data)
    k = int(r*n)
    nbrs = NearestNeighbors(n_neighbors=k, algorithm='ball_tree').fit(data)
    distances, indices = nbrs.kneighbors(data)
    
    sum_distances = []
    for d in distances:
        sum_distances.append(sum(d))
    
    mean_distance = np.mean(sum_distances)
    sd_distance = np.std(sum_distances)
    
    outliers = []
    for i in range(len(sum_distances)):
        if abs(sum_distances[i] - mean_distance) > sd_distance*t:
            outliers.append(data[i])
    return outliers

# parameters 
r = 0.01 # percent of data to use as k for k-nearest neighbors
t = 3 # number of standard deviations away for it to count as an outlier

# plot results
plt.scatter(daily_returns, volume)
plt.xlabel('Daily Return')
plt.ylabel('Volume')
plt.scatter(*zip(*knn_1(data, r, t)), color='red')
plt.legend(['(Return, Volume) points', 'Outliers'], loc = 'center left', bbox_to_anchor = (1.2, 0.12))
plt.grid()
plt.title('Scaled daily volume vs. Scaled daily return for SPY')
plt.show()
```

```python
# k-nearest neighbor algo 2
from sklearn.neighbors import NearestNeighbors
def knn_2(data, k, m, d):
    n = len(data)
    nbrs = NearestNeighbors(n_neighbors=k, algorithm='ball_tree').fit(data)
    distances, indices = nbrs.kneighbors(data)
    
    outliers = []
    for i in range(len(distances)):
        count = 0
        for a in distances[i]:
            if a < d:
                count += 1
        if count <= m:
            outliers.append(data[i])
    return outliers

# parameters 
k = 10
m = 5
d = 1.6

# plot results
plt.scatter(daily_returns, volume)
plt.xlabel('Daily Return')
plt.ylabel('Volume')
plt.scatter(*zip(*knn_2(data, k, m, d)), color='red')
plt.legend(['(Return, Volume) points', 'Outliers'], loc = 'center left', bbox_to_anchor = (1.2, 0.12))
plt.grid()
plt.title('Scaled daily volume vs. Scaled daily return for SPY')
plt.show()
```
